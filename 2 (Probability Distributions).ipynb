```r
# using dnorm and lines to make multiple probability distributions on the same plot

# begin by making an x-vector

x <- seq(0,8      # x will go from zero to 8
         , length.out = 500  # with 500 points
         )

# First, a normal distribution centered on 2 with a wide peak
# all your distributions should be close to zero at 8

# probably a good idea to name your probability vectors according to the cities

city1 <- dnorm(x,     # x is the input vector
                mean = 3, # center
                sd = 1    # width
                )

plot(x,city1  # x and y coordinates                   
     , lwd = 4
     , type = "l"  # draw as lines
     , bty = "n"   # no box
     , yaxt = "n"  # no y-axis 
     , ylab = ""   # no label
     , main = "YOUR TITLE HERE"   # you have to title it
     , xlab = "YOUR TITLE HERE"    # label the axis--INCLUDE UNITS (INCHES) IN THERE
     , col = "royalblue"  # and of course, pretty colors
     
     )


# now, another normal distribution but centered below zero so you just get the decaying part

city2 <- dnorm(x     # x is the input vector
               , mean = 0 # peak is outside the x range, you'll see it falling off
               , sd = 2.5  # very wide peak
                )

# add this distribution using "lines" (which adds a line onto the previous plot)
# you'll need three additional lines for the final plot

lines(x,1*city2  # x and y coordinates
                 # I've included 1 as a scaling factor, adjust when you make your final plot
      , lwd = 4,
      , type = "l"  # draw as lines
      , col = "firebrick"  # STILL MINE YOU CAN'T HAVE IT
    )


# once you have all four lines, label them.  Here's the text code again
# you'll need four labels

text(0,0.35  # location of text
              # NOTE: the y-values will be SMALL, experiment to find them
     , adj = 0 # sets it left-justified.  OPTIONAL
     , col = "firebrick" # THIS IS MINE YOU CAN'T USE IT CHANGE THE COLOR
     , cex = 1 # Character EXpansion--sets the size of the characters.  Default is 1
     , labels = "YOUR TEXT HERE"
     )

# distributions for city3 and city4
# city3 <- 
# city4 <-

# use lines and text to add them to the plot and annotate them
```
