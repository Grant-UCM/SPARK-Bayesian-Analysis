```r
# time-series problem for a stock market index

# x-vector: from 1000 point decrease to 1000 point increase

x <- seq(-1000,1000,1)

# assumption probability: average 25 point increase daily
# about a 200 point range 
beforeMonday <- dcauchy(x
                      , location = 25
                      , scale = 50
                      )

plot(x,beforeMonday
     , main = "DIFFERENT TITLES FOR YOUR PLOTS"
     , xlab = "Daily Change"
     , ylab = ""
     , yaxt = "n"
     , type = "l"
     , lwd = 3
     , bty = "n"
     , xlim = c(-150,150) # although the x-vector is wider, 
                          # we will only look at this range
      )

# Monday evidence:  20 point increase, about the same range as the assumption
# Monday <- dcauchy(


# Tuesday evidence: 75 point increase, about the same range as the assumption
# Tuesday <- dcauchy(


# Wednesday evidence: 100 point decrease, range from about 200 point decrease to no change
# Wednesday <- dnorm(
```
